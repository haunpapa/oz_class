```python
import pandas as pd
```

```python
# Series

data=['a','b,','c','d','e']

se = pd.Series(data)

print(se)
print(se.index)
print(se.values)

se[0:3]

se.name = 'alphabet'
print(se)
```

```python
# DF (DataFrame = Excel)
data ={
    "country" : ['kor','usa','china','japan'],
    "rank" : [1,2,3,4],
    "grade" : ['A','B','C','D']
}

df = pd.DataFrame(data)
```

```python
# 데이터 셀렉션 => 데이터를 불러오는 방법
# 1. df.컬럼, df['컬럼]
df.grade

df[['country','rank']]
```

```python
#2. df.loc[인덱스값,컬럼명]

print(df.loc[3])

print(df.loc[:,['grade','country']])

print(df.loc[df['rank'] >2 ])
```

```python
# 삭제하는 방법 (drop)

# 단순하게 행데이터를 삭제하는 방법
df.drop([3]) # 보여질때만 삭제
# df.drop([3], inplace = True) 완전 삭제
```

```python
df
```

```python
# 컬럼을 삭제
df.drop('rank',axis=1)
```

```python
# 기술 통계
df.info()
df.describe()
```

```python
# 데이터 정렬 => sort_index, sort_values()

df['new_rank']=[10,100,30,60]

df.sort_values(by='new_rank',ascending=True)
#True 오름차순 , False 내림차순

```

```python
# null 데이터 처리 (NaN)
# isnull() => 데이터의 null 여부
# fillna() => null 데이터를 채워주세요
# dropna() => null 데이터 지워주세요
import numpy as np
df.loc[1,'grade'] = np.nan

df.dropna(how='any')
```
